```python
METRICS = "ULB_metrics.pkl"
DATASET_NAME, X_train, X_test, y_train, y_test = joblib.load("ULB_data.pkl")
DE_LONG_RESULTS = joblib.load("ULB_all_results.pkl")
```

```python
DE_LONG_RESULTS.keys()
```

```python
y_test_vec = y_test.iloc[:, 0].values 

REF_MODEL = "BGMM_Gibbs"
ref_scores = DE_LONG_RESULTS[REF_MODEL]["test_scores"]

tables = []
for model_name, info in DE_LONG_RESULTS.items():
    if model_name == REF_MODEL:
        continue
    tables.append(
        delong_auc_validity_table(
            data_name=DATASET_NAME,
            true_labels=y_test_vec,
            model1_scores=ref_scores,
            model2_scores=info["test_scores"],
            model1_name=REF_MODEL,
            model2_name=model_name
        )
    )

final_table = pd.concat(tables, ignore_index=True)
final_table = final_table.sort_values(by=["p_value", "Delta_AUC"], ascending=[True, False]).reset_index(drop=True)

final_table
```
